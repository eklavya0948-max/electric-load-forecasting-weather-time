# Why data cleaning is required?
Answer:
**a**. Our focus in this project is to compare load forecasting of Linear Regression and LSTM using meteorological and temporal factors.
Following features our required for our study:
1. Temprature
2. Humidity
3. Hour
4. Day of week
5. Month

But our dataset contains various other factors such as voltage, current, power factor, weekday, weekend etc. These factors can affect our output accuracy so they must be removed.

**b**. Handling missing values: By examining dataset we can see that there are many missing values which are needed for defining our data. If not handled our dataset isn't consistent and regular. It will affect our models' accuracy.

**c**. Date Time correction: Seperate date time column were combined to single datetime column for proper time series analysis.

**d**.Some time entries used inconsistent formats (e.g., - instead of :). These were standardized to ensure correct datetime conversion.

**e**. Feature Selection:
Only relevant features such as temperature, humidity, seasonal information, and time-based features are kept for modeling.

**f**. Feature Engineering:
Additional temporal features such as hour, day of week, and month were extracted from the datetime column to capture daily and seasonal load patterns.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("2021 load dataset.xlsx")
```

Filling missing data and converting time format

```python
df["DATE"] = df["DATE"].ffill()

df["TIME"] = df["TIME"].str.replace("-", ":")
df["TIME"] = df["TIME"].str.replace("24:00", "00:00")

df["datetime"] = pd.to_datetime(df["DATE"].astype(str) + " " + df["TIME"])

df = df.set_index("datetime")
```

Selecting required features and renaming columns


```python
df = df[["Temp (F)", "Humidity (%)", "POWER (KW)"]]
df.columns = ["temperature","humidity","load"]
```

Extracting temporal data to help model capture load patterns


```python
df["hour"] = df.index.hour
df["day_of_week"] = df.index.dayofweek
df["month"] = df.index.month
```

Saving cleaned dataset as cleaned data.csv

```python
df.to_csv("cleaned data.csv")
```
